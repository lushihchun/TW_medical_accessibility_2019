```python
import json
import geopandas as gpd
import shapely.geometry
import geopandas_osm.osm as goo
import matplotlib.pyplot as plt
import folium
%matplotlib inline
```

```python
village=gpd.read_file("C:\\Users\\susan\\python\\2015村里界\\VA0686.shp",encoding="utf-8")
```

```python
village
```

```python
village["centroid"]=[""]*len(village)
for i in range(len(village)):
    p=village.loc[i,"geometry"].centroid
    village.loc[i,"centroid"]=p
```

```python
village
```

```python
village=gpd.GeoDataFrame(village,geometry="centroid")
```

```python
village=village.drop("geometry",axis=1)
village.to_file("village_centroid.shp",encoding="utf-8")
```

```python
village["經度"]=[""]*len(village)
village["緯度"]=[""]*len(village)
for i in range(len(village)):
    lon=str(village.loc[i,"centroid"]).split("(")[1].split(" ")[0]
    lat=str(village.loc[i,"centroid"]).split("(")[1].split(" ")[1][:-1]
    village.loc[i,"經度"]=lon
    village.loc[i,"緯度"]=lat
```
